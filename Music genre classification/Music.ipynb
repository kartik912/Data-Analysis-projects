# model training and building

```python
df.columns
```

```python
df = df.reset_index(drop = True)
df
```

```python
df2 = df.drop(['Artist Name','Track Name','mode','time_signature','Class','Colaboration'], axis=1)
df2
```

```python
def normalize(feature):
    return (feature-np.mean(feature))/np.std(feature)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df2 = scaler.fit_transform(df2)
```

```python
df2
```

```python
df2 = pd.DataFrame(df2,columns=['Popularity', 'danceability', 'energy',
       'key', 'loudness', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_in min/ms'])

df2
```

```python
sns.kdeplot(df2['key'])
```

```python
# as you can see that the graph is not changes at aloo , just the values decreased which will make the model more efficient
```

```python
df = df.drop(['Popularity', 'danceability', 'energy',
       'key', 'loudness', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_in min/ms'], axis=1)
df
```

```python
df2
```

```python
df = pd.concat([df,df2],axis='columns')
df
```

```python
df = df.drop(['Artist Name','Track Name'], axis=1)
```

```python
df
```

```python
Y = df['Class']
X = df.drop(['Class'], axis=1)
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=10)
```

```python
X_train
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(multi_class='multinomial')

lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

STANDARD VECTOR MACHINE / STANDARD VECTOR CLASSIFIER

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,Y_train)
```

```python
svc.score(X_test,Y_test)
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
rfc.score(X_test,Y_test)
```

K_neighbour classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```

NAIVE BAYES CLASSIFIER

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
gnb.score(X_test,Y_test)
```

```python
from sklearn.model_selection import cross_val_score

models = [lr,svc,knn,gnb]
accuracies=[]
for i in models:
    scores=cross_val_score(i,X,Y,cv=5)
    accuracies.append(np.mean(scores))
accuracies
```
