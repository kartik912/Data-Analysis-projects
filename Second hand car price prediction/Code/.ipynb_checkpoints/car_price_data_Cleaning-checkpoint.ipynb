# Creating dependent and independent features

```python
Y = df['Price']
Y
```

```python
X = df.drop(['Price'],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x,train_y,test_y = train_test_split(X,Y, train_size=0.8,random_state=1)
```

```python
test_x
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_x, train_y)
```

```python
lr_pridected  = lr.predict(test_x)
lr_pridected[:5]
```

```python
test_y[0:5]
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(test_y,lr_pridected)
r2
```

```python
ac = lr.score(test_x, test_y)
```

```python
ac
```

```python
from sklearn.model_selection import cross_val_score
average = np.average(cross_val_score(lr,X,Y,cv=10))
average
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_x,train_y)

ls_pridected  = lasso.predict(test_x)
ls_pridected[:5]
```

```python
r2 = r2_score(test_y,ls_pridected)
r2
```

```python
average1 = np.average(cross_val_score(lasso,X,Y,cv=10))
average1
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(train_x,train_y)

Y_predicted = dtr.predict(test_x)
Y_predicted
```

```python
r2 = r2_score(test_y,Y_predicted)
r2
```

```python
acc= dtr.score(test_x, test_y)
acc
```
