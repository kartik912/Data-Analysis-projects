# Label encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(df['Location'])
```

    Label encoding is the method in which we are it gives different values to each of the Names of location
    
    But the problem with this is that it may relate two values tp each other, we do not need that as we need only independent variables

    SO we will use dummies : 
        What it does is that it will give a binary number type to each of the names of the cities
        
        example : red   - [001]
                  blue  - [010]
                  green - [100]
                  
      It is also known as one-hot encoding in sklearn
    

```python
dummies1 = pd.get_dummies(df['Location'])
dummies1
```

```python
df = df.drop(['Location'], axis='columns')
df = pd.concat([df,dummies1], axis=1)
```

```python
df
```
